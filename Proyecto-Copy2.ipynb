```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("keplersmachines/kepler-labelled-time-series-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Buscar el CSV dentro de la ruta que devolviÃ³ kagglehub
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".csv"):
            csv_path = os.path.join(root, file)
            print("Archivo CSV encontrado:", csv_path)

# Cargar el dataset
deep_space = pd.read_csv(csv_path)
deep_space.head()
```
